#### Q1. 1. Merge the datasets and visualize the data by considering its quality label 

```python
#adding a column to indicate colour when the dfs are merged
#number of entries is now total - missing values (removed)

white = np.repeat('white', 3961)
red = np.repeat('red', 1359)
df1['color'] = red
df2['color'] = white
```

```python
#appending the rows of the two dfs (merging)

winedf = df1.append(df2, ignore_index=True)
winedf
```

```python
#visualising the data based on quality level
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
ax = winedf['quality'].plot(kind = "hist")
```

## Q2. Sulphates and alcohol

```python
df3 = winedf[['sulphates', 'alcohol']]
sns.pairplot(df3, height=10, aspect=1)
```

```python
df3.corr(method = 'pearson')
```

negative correlation, i.e, higher the sulphates, lower the alcohol.

## Q3. Correlation between various attributes

```python
#correlation between various attributes

df4 = winedf[['density', 'total sulfur dioxide', 'residual sugar', 'fixed acidity']]
plt.figure(figsize=(12, 10))
ax = sns.heatmap(df4.corr(), annot=True, fmt='.3g')
```

```python
plt.figure(figsize=(10, 15))
ax = sns.pairplot(df4)
```
