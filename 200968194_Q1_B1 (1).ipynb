```python
# IMPORTING LIBRARIES
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# READING FILE
df=pd.read_csv(r"C:\Users\nidaj\Downloads\diabetes.csv")
df
```

```python
# Q1)What is the percentage of patients who tested positive for diabetes.

total = df.shape[0]
no_of_pep = df[df["class"] == "tested_positive"].shape[0]
percentage_pep = (no_of_pep/total)*100
percentage_pep
```

```python
# Q2)Analyse  the  pedigree  function  vs  class  (diabetic  or  not).  Visualize  using  appropriate plot.

df.boxplot(by="class" , column=["pedi"])
```

```python
# Q3)Tabulate the Diastolic blood pressure with class (Diabetic or not). 

df['risk']=pd.cut(x=df['pres'], bins=[-1,80,89,200],labels = ['Normal', 'At Risk', 'High Blood Pressure'])
df
```
